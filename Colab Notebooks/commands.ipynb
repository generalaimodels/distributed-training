## 7. CONTROL FLOW & LOOP CONSTRUCTS

```powershell
# ── If / ElseIf / Else ──
if ($value -gt 100) {
    Write-Output "High"
} elseif ($value -gt 50) {
    Write-Output "Medium"
} else {
    Write-Output "Low"
}

# ── Switch ──
switch ($status) {
    "Running"   { Write-Output "Active" }
    "Stopped"   { Write-Output "Inactive" }
    "Degraded"  { Write-Output "Warning" }
    default     { Write-Output "Unknown" }
}

switch -Regex ($input) {
    "^\d+$"     { "Number: $_" }
    "^[a-z]+$"  { "Lowercase: $_" }
    "^[A-Z]+$"  { "Uppercase: $_" }
}

switch -Wildcard ($path) {
    "*.log"  { "Log file" }
    "*.json" { "JSON file" }
    "*.xml"  { "XML file" }
}

switch -File "C:\data\list.txt" {
    { $_ -match "ERROR" } { "Error line: $_" }
    { $_ -match "WARN" }  { "Warning line: $_" }
}

# ── For Loop ──
for ($i = 0; $i -lt 10; $i++) {
    Write-Output $i
}

# ── ForEach Loop ──
foreach ($item in $collection) {
    Write-Output $item
}

# ── ForEach-Object (Pipeline) ──
1..10 | ForEach-Object { $_ * 2 }
1..10 | ForEach-Object -Begin { $sum = 0 } -Process { $sum += $_ } -End { $sum }
1..100 | ForEach-Object -Parallel { $_ * 2 } -ThrottleLimit 10   # PS 7+

# ── While ──
$i = 0
while ($i -lt 10) {
    Write-Output $i
    $i++
}

# ── Do-While ──
$i = 0
do {
    Write-Output $i
    $i++
} while ($i -lt 10)

# ── Do-Until ──
$i = 0
do {
    Write-Output $i
    $i++
} until ($i -ge 10)

# ── Loop Control ──
foreach ($item in 1..20) {
    if ($item -eq 5) { continue }                    # Skip iteration
    if ($item -eq 15) { break }                      # Exit loop
    Write-Output $item
}

# ── Labels ──
:outer foreach ($i in 1..5) {
    foreach ($j in 1..5) {
        if ($j -eq 3) { continue outer }
        if ($i -eq 4) { break outer }
        Write-Output "$i-$j"
    }
}
```

```powershell
# ── Basic Function ──
function Get-Greeting {
    param([string]$Name = "World")
    return "Hello, $Name"
}

# ── Advanced Function (Production Grade) ──
function Invoke-DeploymentValidation {
    [CmdletBinding(SupportsShouldProcess, ConfirmImpact = 'High')]
    [OutputType([PSCustomObject])]
    param(
        [Parameter(Mandatory, Position = 0, ValueFromPipeline, ValueFromPipelineByPropertyName)]
        [ValidateNotNullOrEmpty()]
        [Alias("Server")]
        [string[]]$ComputerName,

        [Parameter(Mandatory)]
        [ValidateSet("Production", "Staging", "Development")]
        [string]$Environment,

        [Parameter()]
        [ValidateRange(1, 65535)]
        [int]$Port = 443,

        [Parameter()]
        [ValidatePattern("^v\d+\.\d+\.\d+$")]
        [string]$Version = "v1.0.0",

        [Parameter()]
        [ValidateScript({ Test-Path $_ -PathType Leaf })]
        [string]$ConfigPath,

        [Parameter()]
        [ValidateLength(1, 256)]
        [string]$Description,

        [Parameter()]
        [ValidateCount(1, 10)]
        [string[]]$Tags,

        [Parameter()]
        [ValidateNotNull()]
        [PSCredential]$Credential,

        [Parameter()]
        [switch]$Force,

        [Parameter()]
        [int]$TimeoutSeconds = 30
    )

    begin {
        Write-Verbose "Initializing deployment validation pipeline"
        $results = [System.Collections.Generic.List[PSCustomObject]]::new()
    }

    process {
        foreach ($computer in $ComputerName) {
            if ($PSCmdlet.ShouldProcess($computer, "Validate deployment")) {
                try {
                    Write-Verbose "Validating $computer"
                    $result = [PSCustomObject]@{
                        ComputerName = $computer
                        Environment  = $Environment
                        Port         = $Port
                        Version      = $Version
                        Status       = "Validated"
                        Timestamp    = Get-Date -Format "o"
                    }
                    $results.Add($result)
                    Write-Output $result
                }
                catch {
                    Write-Error "Validation failed for ${computer}: $_"
                }
            }
        }
    }

    end {
        Write-Verbose "Validation complete. Processed $($results.Count) hosts."
    }
}

# ── Parameter Sets ──
function Get-SystemInfo {
    [CmdletBinding(DefaultParameterSetName = "ByName")]
    param(
        [Parameter(ParameterSetName = "ByName", Mandatory)]
        [string]$Name,

        [Parameter(ParameterSetName = "ById", Mandatory)]
        [int]$Id,

        [Parameter(ParameterSetName = "All")]
        [switch]$All
    )

    switch ($PSCmdlet.ParameterSetName) {
        "ByName" { "Fetching by name: $Name" }
        "ById"   { "Fetching by ID: $Id" }
        "All"    { "Fetching all" }
    }
}

# ── Dynamic Parameters ──
function Get-DynamicExample {
    [CmdletBinding()]
    param([string]$Type)

    dynamicparam {
        if ($Type -eq "Advanced") {
            $paramDictionary = [System.Management.Automation.RuntimeDefinedParameterDictionary]::new()
            $attr = [System.Management.Automation.ParameterAttribute]@{ Mandatory = $true }
            $attrCollection = [System.Collections.ObjectModel.Collection[System.Attribute]]::new()
            $attrCollection.Add($attr)
            $dynParam = [System.Management.Automation.RuntimeDefinedParameter]::new(
                "DetailLevel", [int], $attrCollection
            )
            $paramDictionary.Add("DetailLevel", $dynParam)
            return $paramDictionary
        }
    }

    process {
        $PSBoundParameters
    }
}

# ── Splatting ──
$params = @{
    Path        = "C:\Logs"
    Filter      = "*.log"
    Recurse     = $true
    ErrorAction = "SilentlyContinue"
}
Get-ChildItem @params

# ── Argument Completers ──
Register-ArgumentCompleter -CommandName Get-SystemInfo -ParameterName Name -ScriptBlock {
    param($commandName, $parameterName, $wordToComplete, $commandAst, $fakeBoundParameters)
    @("Server01", "Server02", "Server03") | Where-Object { $_ -like "$wordToComplete*" } |
        ForEach-Object { [System.Management.Automation.CompletionResult]::new($_) }
}

# ── Filter Function ──
filter Get-LargeFile {
    if ($_.Length -gt 1MB) { $_ }
}
Get-ChildItem -Recurse | Get-LargeFile

# ── Begin/Process/End Pattern ──
function Measure-Pipeline {
    [CmdletBinding()]
    param([Parameter(ValueFromPipeline)][object]$InputObject)
    begin   { $count = 0; $sw = [System.Diagnostics.Stopwatch]::StartNew() }
    process { $count++ }
    end     { $sw.Stop(); [PSCustomObject]@{ Count = $count; ElapsedMs = $sw.ElapsedMilliseconds } }
}
```

```powershell
# ── Try/Catch/Finally ──
try {
    $result = Get-Content -Path "nonexistent.txt" -ErrorAction Stop
}
catch [System.IO.FileNotFoundException] {
    Write-Error "File not found: $($_.Exception.Message)"
}
catch [System.UnauthorizedAccessException] {
    Write-Error "Access denied: $($_.Exception.Message)"
}
catch [System.Management.Automation.ItemNotFoundException] {
    Write-Error "Item not found: $($_.Exception.Message)"
}
catch {
    Write-Error "Unexpected error: $($_.Exception.GetType().FullName) - $($_.Exception.Message)"
}
finally {
    Write-Verbose "Cleanup complete"
}

# ── Error Record Inspection ──
$Error[0]
$Error[0].Exception
$Error[0].Exception.Message
$Error[0].Exception.GetType().FullName
$Error[0].Exception.InnerException
$Error[0].InvocationInfo
$Error[0].InvocationInfo.ScriptLineNumber
$Error[0].InvocationInfo.Line
$Error[0].InvocationInfo.PositionMessage
$Error[0].ScriptStackTrace
$Error[0].CategoryInfo
$Error[0].FullyQualifiedErrorId
$Error[0].TargetObject

# ── ErrorAction Parameter ──
Get-Item "missing" -ErrorAction Stop
Get-Item "missing" -ErrorAction SilentlyContinue
Get-Item "missing" -ErrorAction Continue
Get-Item "missing" -ErrorAction Ignore
Get-Item "missing" -ErrorAction Inquire

# ── Error Variable ──
Get-Item "missing" -ErrorAction SilentlyContinue -ErrorVariable capturedError
$capturedError

# ── Throw ──
throw "Critical failure occurred"
throw [System.InvalidOperationException]::new("Operation not permitted")

# ── Write-Error ──
Write-Error "Non-terminating error" -Category InvalidArgument -ErrorId "INVALID_ARG"
Write-Error -Exception ([System.IO.IOException]::new("IO failure")) -ErrorId "IO_ERR"

# ── Trap ──
trap {
    Write-Error "Trapped: $_"
    continue                                          # or break
}

# ── Custom Exception ──
class AutomationException : System.Exception {
    [string]$Component
    AutomationException([string]$message, [string]$component) : base($message) {
        $this.Component = $component
    }
}
throw [AutomationException]::new("Pipeline failed", "Deployer")

# ── Retry Pattern ──
function Invoke-WithRetry {
    [CmdletBinding()]
    param(
        [Parameter(Mandatory)]
        [scriptblock]$ScriptBlock,
        [int]$MaxRetries = 3,
        [int]$DelaySeconds = 2,
        [double]$BackoffMultiplier = 2.0
    )
    $attempt = 0
    $delay = $DelaySeconds
    while ($true) {
        $attempt++
        try {
            return & $ScriptBlock
        }
        catch {
            if ($attempt -ge $MaxRetries) {
                Write-Error "All $MaxRetries attempts failed. Last error: $_"
                throw
            }
            Write-Warning "Attempt $attempt/$MaxRetries failed: $($_.Exception.Message). Retrying in ${delay}s..."
            Start-Sleep -Seconds $delay
            $delay = [math]::Round($delay * $BackoffMultiplier)
        }
    }
}
```

## 34. CI/CD PIPELINE SCRIPTING

```powershell
# ── Build Script Pattern ──
#Requires -Version 7.0
$ErrorActionPreference = "Stop"
Set-StrictMode -Version Latest

function Invoke-Build {
    [CmdletBinding()]
    param(
        [ValidateSet("Debug","Release")]
        [string]$Configuration = "Release",
        [string]$OutputPath = "$PSScriptRoot\artifacts",
        [switch]$Clean
    )

    $sw = [System.Diagnostics.Stopwatch]::StartNew()

    try {
        if ($Clean -and (Test-Path $OutputPath)) {
            Write-Output ":: Cleaning $OutputPath"
            Remove-Item $OutputPath -Recurse -Force
        }
        New-Item -Path $OutputPath -ItemType Directory -Force | Out-Null

        Write-Output ":: Restoring dependencies"
        & dotnet restore --no-cache
        if ($LASTEXITCODE -ne 0) { throw "Restore failed" }

        Write-Output ":: Building ($Configuration)"
        & dotnet build --configuration $Configuration --output $OutputPath --no-restore
        if ($LASTEXITCODE -ne 0) { throw "Build failed" }

        Write-Output ":: Running tests"
        & dotnet test --configuration $Configuration --no-build --logger "trx;LogFileName=results.trx" --results-directory "$OutputPath\TestResults"
        if ($LASTEXITCODE -ne 0) { throw "Tests failed" }

        Write-Output ":: Packaging"
        & dotnet publish --configuration $Configuration --output "$OutputPath\publish" --no-build
        if ($LASTEXITCODE -ne 0) { throw "Publish failed" }

        $sw.Stop()
        Write-Output ":: Build completed in $([math]::Round($sw.Elapsed.TotalSeconds, 2))s"
    }
    catch {
        $sw.Stop()
        Write-Error ":: BUILD FAILED after $([math]::Round($sw.Elapsed.TotalSeconds, 2))s: $_"
        exit 1
    }
}

# ── Idempotent Deployment ──
function Invoke-Deployment {
    [CmdletBinding(SupportsShouldProcess)]
    param(
        [Parameter(Mandatory)]
        [string]$ArtifactPath,
        [Parameter(Mandatory)]
        [string]$TargetPath,
        [string]$BackupPath = "$TargetPath.backup.$(Get-Date -Format 'yyyyMMddHHmmss')"
    )

    if (-not (Test-Path $ArtifactPath)) { throw "Artifact not found: $ArtifactPath" }

    try {
        if (Test-Path $TargetPath) {
            Write-Output ":: Creating backup: $BackupPath"
            Copy-Item -Path $TargetPath -Destination $BackupPath -Recurse -Force
        }

        if ($PSCmdlet.ShouldProcess($TargetPath, "Deploy from $ArtifactPath")) {
            Write-Output ":: Deploying to $TargetPath"
            Copy-Item -Path "$ArtifactPath\*" -Destination $TargetPath -Recurse -Force

            Write-Output ":: Validating deployment"
            if (-not (Test-Path "$TargetPath\app.dll")) {
                throw "Validation failed: app.dll missing"
            }
            Write-Output ":: Deployment successful"
        }
    }
    catch {
        Write-Error ":: Deployment failed: $_"
        if (Test-Path $BackupPath) {
            Write-Warning ":: Rolling back to $BackupPath"
            Remove-Item $TargetPath -Recurse -Force -ErrorAction SilentlyContinue
            Move-Item $BackupPath $TargetPath -Force
            Write-Output ":: Rollback complete"
        }
        throw
    }
}

# ── Health Check ──
function Test-ServiceHealth {
    [CmdletBinding()]
    param(
        [string]$Endpoint,
        [int]$ExpectedStatus = 200,
        [int]$TimeoutSeconds = 10,
        [int]$RetryCount = 3,
        [int]$RetryDelaySeconds = 5
    )
    for ($i = 1; $i -le $RetryCount; $i++) {
        try {
            $response = Invoke-WebRequest -Uri $Endpoint -TimeoutSec $TimeoutSeconds -UseBasicParsing -ErrorAction Stop
            if ($response.StatusCode -eq $ExpectedStatus) {
                Write-Output "Health check passed (attempt $i)"
                return $true
            }
        }
        catch {
            Write-Warning "Health check attempt $i/$RetryCount failed: $_"
            if ($i -lt $RetryCount) { Start-Sleep $RetryDelaySeconds }
        }
    }
    throw "Health check failed after $RetryCount attempts"
}
```

## 36. PERFORMANCE PROFILING & OPTIMIZATION

```powershell
# ── Measure-Command ──
Measure-Command { 1..100000 | ForEach-Object { $_ * 2 } }
Measure-Command { for ($i = 0; $i -lt 100000; $i++) { $i * 2 } }    # Faster
Measure-Command { [System.Linq.Enumerable]::Select([int[]](1..100000), [func[int,int]]{ param($x) $x * 2 }) }

# ── Trace-Command ──
Trace-Command -Name ParameterBinding -Expression { Get-Process -Name "svchost" } -PSHost
Trace-Command -Name TypeConversion -Expression { [int]"42" } -PSHost

# ── .NET Stopwatch ──
$sw = [System.Diagnostics.Stopwatch]::StartNew()
# ... code ...
$sw.Stop()
"Elapsed: $($sw.ElapsedMilliseconds)ms"

# ── Memory Measurement ──
[System.GC]::Collect()
$before = [System.GC]::GetTotalMemory($true)
# ... allocate ...
$after = [System.GC]::GetTotalMemory($true)
"Memory delta: $(($after - $before) / 1KB) KB"

# ── Performance Best Practices ──
# AVOID: += on arrays (copies entire array each time)
$arr = @()
$arr += "item"                                        # O(n²) total

# USE: Generic List
$list = [System.Collections.Generic.List[string]]::new()
$list.Add("item")                                     # O(1) amortized

# USE: StringBuilder for string concatenation
$sb = [System.Text.StringBuilder]::new()
[void]$sb.Append("chunk")

# USE: for loops over ForEach-Object for CPU-bound
for ($i = 0; $i -lt $arr.Count; $i++) { $arr[$i] }

# USE: .Where() and .ForEach() methods over pipeline for in-memory collections
(1..100000).Where({ $_ -gt 50000 })
(1..100000).ForEach({ $_ * 2 })

# AVOID: Write-Host in loops (slow)
# USE: [Console]::WriteLine() for raw speed

# USE: -Filter parameter instead of Where-Object post-filter
Get-ChildItem -Filter "*.log"                         # Fast (provider-level)
Get-ChildItem | Where-Object Extension -EQ ".log"     # Slow (PowerShell-level)

# USE: Select-Object -First for early termination
Get-ChildItem -Recurse | Select-Object -First 1

# SUPPRESS progress bar
$ProgressPreference = "SilentlyContinue"

# ── Pipeline vs Method Performance ──
# Pipeline: convenient, slower for large datasets
1..1000000 | Where-Object { $_ % 2 -eq 0 } | Measure-Object

# Method: faster for in-memory operations
$data = 1..1000000
$data.Where({ $_ % 2 -eq 0 }).Count

# ── GC Control ──
[System.GC]::Collect()
[System.GC]::WaitForPendingFinalizers()
[System.GC]::GetTotalMemory($false)
```

## 48. CONFIGURATION-AS-CODE PATTERN

```powershell
# ── Configuration Management ──
class AppConfig {
    [string]$Environment
    [hashtable]$Database
    [hashtable]$Logging
    [hashtable]$Features
    [string]$Version
    hidden [datetime]$LoadedAt

    static [AppConfig] Load([string]$path) {
        if (-not (Test-Path $path)) { throw "Config not found: $path" }
        $raw = Get-Content $path -Raw | ConvertFrom-Json
        $config = [AppConfig]::new()
        $config.Environment = $raw.Environment
        $config.Database = @{
            Server   = $raw.Database.Server
            Name     = $raw.Database.Name
            Port     = $raw.Database.Port
            Timeout  = $raw.Database.Timeout
        }
        $config.Logging = @{
            Level = $raw.Logging.Level
            Path  = $raw.Logging.Path
        }
        $config.Features = @{}
        foreach ($prop in $raw.Features.PSObject.Properties) {
            $config.Features[$prop.Name] = $prop.Value
        }
        $config.Version = $raw.Version
        $config.LoadedAt = Get-Date
        return $config
    }

    [void] Save([string]$path) {
        $this | ConvertTo-Json -Depth 10 | Set-Content $path -Encoding UTF8
    }

    [void] Validate() {
        if ([string]::IsNullOrWhiteSpace($this.Environment)) { throw "Environment required" }
        if (-not $this.Database.Server) { throw "Database server required" }
        if ($this.Database.Port -lt 1 -or $this.Database.Port -gt 65535) { throw "Invalid port" }
    }
}

# ── Environment-Specific Config Loading ──
function Get-EnvironmentConfig {
    param([string]$Environment = $env:APP_ENVIRONMENT ?? "Development")
    $configPath = Join-Path $PSScriptRoot "config.$Environment.json"
    $config = [AppConfig]::Load($configPath)
    $config.Validate()
    return $config
}
```

## 50. PRODUCTION SCRIPT TEMPLATE

```powershell
#Requires -Version 7.0
#Requires -Modules @{ ModuleName="Pester"; ModuleVersion="5.0" }

<#
.SYNOPSIS
    Production-grade automation script template.
.DESCRIPTION
    Implements enterprise patterns: structured logging, error handling,
    retry logic, configuration management, and observability.
.PARAMETER Environment
    Target deployment environment.
.PARAMETER ConfigPath
    Path to JSON configuration file.
.PARAMETER DryRun
    Execute in simulation mode without side effects.
.EXAMPLE
    .\Invoke-Automation.ps1 -Environment Production -ConfigPath .\config.json
.NOTES
    Author: Automation Engineering
    Version: 1.0.0
    Date: 2025-01-15
#>

[CmdletBinding(SupportsShouldProcess)]
param(
    [Parameter(Mandatory)]
    [ValidateSet("Production", "Staging", "Development")]
    [string]$Environment,

    [Parameter()]
    [ValidateScript({ Test-Path $_ -PathType Leaf })]
    [string]$ConfigPath = "$PSScriptRoot\config.$Environment.json",

    [Parameter()]
    [switch]$DryRun
)

Set-StrictMode -Version Latest
$ErrorActionPreference = "Stop"
$ProgressPreference = "SilentlyContinue"

#region Initialization
$script:LogPath = Join-Path $PSScriptRoot "logs" "$(Get-Date -Format 'yyyyMMdd_HHmmss')_$Environment.log"
New-Item -Path (Split-Path $script:LogPath) -ItemType Directory -Force | Out-Null
$script:Stopwatch = [System.Diagnostics.Stopwatch]::StartNew()
$script:ExitCode = 0
#endregion

#region Functions
function Write-Log {
    param(
        [string]$Message,
        [ValidateSet("INFO","WARN","ERROR","DEBUG","FATAL")]
        [string]$Level = "INFO"
    )
    $entry = "[{0:yyyy-MM-ddTHH:mm:ss.fffzzz}] [{1}] [{2}] {3}" -f (Get-Date), $Level, $Environment, $Message
    Add-Content -Path $script:LogPath -Value $entry -Encoding UTF8
    switch ($Level) {
        "ERROR" { Write-Error $entry -ErrorAction Continue }
        "WARN"  { Write-Warning $entry }
        "FATAL" { Write-Error $entry -ErrorAction Continue }
        default { Write-Verbose $entry }
    }
}

function Invoke-WithRetry {
    param(
        [scriptblock]$Action,
        [string]$Description = "operation",
        [int]$MaxAttempts = 3,
        [int]$BaseDelaySeconds = 2
    )
    for ($attempt = 1; $attempt -le $MaxAttempts; $attempt++) {
        try {
            Write-Log "Executing $Description (attempt $attempt/$MaxAttempts)"
            return (& $Action)
        }
        catch {
            Write-Log "$Description failed (attempt $attempt): $($_.Exception.Message)" -Level WARN
            if ($attempt -eq $MaxAttempts) { throw }
            $delay = $BaseDelaySeconds * [math]::Pow(2, $attempt - 1)
            Start-Sleep -Seconds $delay
        }
    }
}
#endregion

#region Main
try {
    Start-Transcript -Path ($script:LogPath -replace '\.log$', '_transcript.log') -Append
    Write-Log "=== Automation started ==="
    Write-Log "Environment: $Environment | DryRun: $DryRun | Host: $env:COMPUTERNAME"
    Write-Log "PowerShell: $($PSVersionTable.PSVersion) ($($PSVersionTable.PSEdition))"

    # Load configuration
    Write-Log "Loading configuration from $ConfigPath"
    $config = Get-Content $ConfigPath -Raw | ConvertFrom-Json

    # Execute operations
    if ($DryRun) {
        Write-Log "DRY RUN: No changes will be made" -Level WARN
    }

    # ... automation logic ...

    Write-Log "=== Automation completed successfully ==="
}
catch {
    $script:ExitCode = 1
    Write-Log "FATAL: $($_.Exception.Message)" -Level FATAL
    Write-Log "Stack: $($_.ScriptStackTrace)" -Level FATAL
    throw
}
finally {
    $script:Stopwatch.Stop()
    Write-Log "Total execution time: $($script:Stopwatch.Elapsed.ToString('hh\:mm\:ss\.fff'))"
    Stop-Transcript -ErrorAction SilentlyContinue
    exit $script:ExitCode
}
#endregion
```
